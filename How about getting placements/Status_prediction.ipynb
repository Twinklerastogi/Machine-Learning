# Status Prediction for Students

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('Placement_Data_Full_Class.csv')
```

```python
data['salary'].fillna(0, inplace = True)
```

```python
columns = ['gender', 'ssc_p', 'ssc_b', 'hsc_b', 'hsc_p', 'degree_t']
data_x = data[columns]
```

```python
data_x.info()
```

### Category values to numerical values (conversion)

```python
def cat_to_num(data_x, col):
    dummy = pd.get_dummies(data_x[col])
    del dummy[dummy.columns[-1]] #To avoid dummy variable
    data_x = pd.concat([data_x, dummy], axis = 1)
    return data_x
```

```python
for i in data_x.columns:
    if data_x[i].dtype==object:
        print(i)
    data_x  = cat_to_num(data_x, i)
```

```python
data_x.drop(['gender', 'ssc_b', 'hsc_b', 'degree_t'], inplace = True, axis = 1)
```

```python
le=LabelEncoder()
data['status'] = le.fit_transform(data['status'])
```

```python
y=data['status']
x=data_x
```

```python
x.info()
```

```python
y.value_counts()
```

### Splitting data into training and testing sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

## Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier as DTC
model = DTC()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,model.predict(X_test)))
```

Our model is 83 percent accurate.

```python
model.predict(X_test)
```
